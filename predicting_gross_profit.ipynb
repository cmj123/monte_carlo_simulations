```python
# Expected revenue  = $170m with a standard deviation of $20m

rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
# Expected COGS ~ 60% of the revenues
COGS = - (rev * np.random.normal(0.6, 0.01))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```
